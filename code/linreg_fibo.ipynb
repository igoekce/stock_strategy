```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import requests
import pandas_ta as ta
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
def prediction(index):
  data_orig = yf.download(index)
  print("Index Name: ", index)

  #Linear Regression
  data = data_orig.reset_index()
  df = data[(4*-365):]
  x = np.array(df.index).reshape(-1, 1)
  y = df['Close']
# to do: test
  linreg= LinearRegression().fit(x,y)
  linreg.score(x, y)
  predictions = linreg.predict(x)

  #plt.ticklabel_format(style='plain')
  plt.figure(figsize=(15,5))
  plt.plot(df['Close'])
  plt.plot(df.index, predictions)

  print('R^2:', linreg.score(x, y))

  print("Predictions", predictions)

  print("min date f√ºr ", data_orig.index[(4*-365):].min())
  long_rolling_200t = data_orig.Close.rolling(window=200).mean()
  long_rolling_200w = data_orig.Close.rolling(window=1400).mean()
  print("200 Tage linie liegt bei.       : ", round(long_rolling_200t.tail().iloc[-1],2))
  print("200 Wochen linie liegt bei.       : ", round(long_rolling_200w.tail().iloc[-1],2))
  print("Regressionsvorhersage liegt bei : ", round(predictions[-1],2))
  stock = yf.Ticker(index)
  print("Current Market Price: ", stock.info['regularMarketPrice'])

  print()
  print('##########')
  print()


  print("Average Volume by YAhoo Finance stock.info: ", stock.info['averageVolume'])
  #print("Derzeit liegt der Dax bei       : ", df.iloc[0].loc['Akt.'])
  df = data[(2*-365):]
  print("Min Date for Volume calculation: ", data_orig.index.min())
  print("Volume average since min Date: ", round(df['Volume'].mean(), 0))

  #df['Volume'].plot()
  #del df

  #Fibo Retracement
  #Calculate the max and min close price
  df = data[(2*-365):]
  print("Rows in df for date: ", df.Close.count())
  maximum_price = df['Close'].max()
  minimum_price = df['Close'].min()
  difference = maximum_price - minimum_price #Get the difference
  first_level = maximum_price - difference * 0.236
  second_level = maximum_price - difference * 0.382
  third_level = maximum_price - difference * 0.5
  fourth_level = maximum_price - difference * 0.618

  print()
  print('##########')
  print()


  #Plot the Fibonacci levels along with the close price
  new_df = df
  plt.figure(figsize=(12.33,4.5))
  plt.title('Fibonnacci Retracement Plot')
  plt.plot(new_df.index, new_df['Close'])
  plt.axhline(maximum_price, linestyle='--', alpha=0.5, color = 'red')
  plt.axhline(first_level, linestyle='--', alpha=0.5, color = 'orange')
  plt.axhline(second_level, linestyle='--', alpha=0.5, color = 'yellow')
  plt.axhline(third_level, linestyle='--', alpha=0.5, color = 'green')
  plt.axhline(fourth_level, linestyle='--', alpha=0.5, color = 'blue')
  plt.axhline(minimum_price, linestyle='--', alpha=0.5, color = 'purple')
  plt.xlabel('Date',fontsize=18)
  plt.ylabel('Close Price in USD',fontsize=18)
  plt.show()

  #Print the price at each level
  print("Level Percentage\t", "Price ($)")
  print("00.0%\t\t", maximum_price)
  print("23.6%\t\t", first_level)
  print("38.2%\t\t", second_level)
  print("50.0%\t\t", third_level)
  print("61.8%\t\t", fourth_level)
  print("100.0%\t\t", minimum_price)
  print('###################################################################')

######################################################################################

indices = ['^GDAXI', '^DJI', '^IXIC', '^GSPC', '^RUT', '^STOXX50E', '^N225', '^HSI',]
index = '^GDAXI'
#prediction(index)

for index in indices:
  prediction(index)
```
